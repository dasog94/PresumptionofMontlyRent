```python
import pandas as pd
```

```python
df_real = pd.read_csv('adjusted_주변_연립다세대(전월세)_실거래가_ALL.csv', encoding='CP949')

df_real = df_real.rename(columns={'시군구': '법정동명', '번지':'지번', '층':'계약층'})

col_list = ['법정동명', '지번', '전월세구분', '전용면적(㎡)', '계약년월', '보증금(만원)', '월세(만원)', '계약층', '건축년도', '도로명']
df_real = df_real.loc[:, col_list]


print('adjusted_주변_연립다세대(전월세)_실거래가_ALL')
print('')
print(df_real.columns)
```

```python
# 대상 csv
df_target = pd.read_csv('df_target_v3.csv', encoding='CP949')

print('대상 csv')
print('')
print(df_target.columns)
```

```python
df_merged = df_real.merge(df_target, how='left', on=['법정동명', '지번'], suffixes=('','_가격정보'))
```

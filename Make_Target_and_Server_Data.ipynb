# Data_Extract

가격정보

```python
# 법정동 리스트 만들기
jong_regions_list = ['혜화동', '명륜1가', '명륜2가', '명륜3가', '명륜4가', '동숭동', '이화동']
sung_regions_list = ['성북동1가', '삼선동1가', '동소문동1가', '동소문동2가', '동소문동3가', '동소문동4가', '동소문동5가', '동소문동6가', '동선동1가', '동선동2가', '동선동3가', '동선동4가']

for i in range(len(jong_regions_list)):
    jong_regions_list[i] = '서울특별시 종로구 ' + jong_regions_list[i]

for i in range(len(sung_regions_list)):
    sung_regions_list[i] = '서울특별시 성북구 ' + sung_regions_list[i]
    
regions_list =  jong_regions_list + sung_regions_list

# 개별주택가격정보
# 지역, 2020년도 기준으로 데이터 추출
df_individual = df_individual[ df_individual['법정동명'].isin(regions_list) ]
df_individual = df_individual[ df_individual['기준년도'] == 2020 ]

# # 필요한 칼럼만 추출
individual_cols = ['법정동명', '지번', '건물산정연면적', '주택가격']
df_individual = df_individual.loc[:, individual_cols]
for i in range(len(df_individual)):
    df_individual['주택구분명'] = '단독다가구'

# 공동주택가격정보
# 지역, 2020년도 기준으로 데이터 추출
df_public = df_public[ df_public['법정동명'].isin(regions_list) ]
df_public = df_public[ df_public['기준년도'] == 2020 ]
df_public = df_public[ df_public['전용면적'] <= 40 ]

# # 필요한 칼럼만 추출
public_cols = ['법정동명', '지번', '공동주택구분명', '전용면적', '공시가격']
df_public = df_public.loc[:, public_cols]
df_public = df_public.rename(columns={'공동주택구분명': '주택구분명'})
```

용도별건물정보

```python
# 법정동 리스트 만들기
jong_regions_list = ['혜화동', '명륜1가', '명륜2가', '명륜3가', '명륜4가', '동숭동', '이화동']
sung_regions_list = ['성북동1가', '삼선동1가', '동소문동1가', '동소문동2가', '동소문동3가', '동소문동4가', '동소문동5가', '동소문동6가', '동선동1가', '동선동2가', '동선동3가', '동선동4가']

for i in range(len(jong_regions_list)):
    jong_regions_list[i] = '서울특별시 종로구 ' + jong_regions_list[i]

for i in range(len(sung_regions_list)):
    sung_regions_list[i] = '서울특별시 성북구 ' + sung_regions_list[i]
    
regions_list =  jong_regions_list + sung_regions_list

# 지역 기준으로 데이터 추출
df_sung = df_sung[ df_sung['법정동명'].isin(regions_list) ]
df_jong = df_jong[ df_jong['법정동명'].isin(regions_list) ]

# 필요한 칼럼만 추출
info_cols = ['법정동명', '지번', '세부용도명', '지하층수', '지상층수', '사용승인일자']

df_sung = df_sung.loc[:, info_cols]
df_jong = df_jong.loc[:, info_cols]

df_sung.reset_index(drop=True, inplace=True)
df_jong.reset_index(drop=True, inplace=True)
```

# Add 6평 가격 Column

```python
# 6평 가격 추가
def Make_STD_Peong_col(df, ind):
    peong = 20 # 원래 19.83 
    for i in range(len(df)):
        if ind:
            df.loc[i, "6평_가격"] = round( df.loc[i, "주택가격"] / df.loc[i, "건물산정연면적"] * peong )
        else:
            df.loc[i, "6평_가격"] = round( df.loc[i, "공시가격"] / df.loc[i, "전용면적"] * peong )

Make_STD_Peong_col(df_individual, True)
Make_STD_Peong_col(df_public, False)
```

# 결합

건물가격정보

```python
# 서로 다른 칼럼이 존재하지만, 일단 합쳐서 preprocess 단계에서 제거할 계획
ind_not_col_list = []
pub_not_col_list = []

for each in df_public.columns:
    if each not in df_individual.columns:
        ind_not_col_list.append(each)
        
for each in df_individual.columns:
    if each not in df_public.columns:
        pub_not_col_list.append(each)
        

df_individual = df_individual.reindex(columns = df_individual.columns.tolist() + ind_not_col_list)
df_public = df_public.reindex(columns = df_public.columns.tolist() + pub_not_col_list)


#결합
df_concated = pd.concat([df_individual, df_public], axis=0)
df_concated.reset_index(drop=True, inplace=True)
```

```python
total = df_concated.isnull().sum().sort_values(ascending=False)
percent = (df_concated.isnull().sum()/df_concated.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```
